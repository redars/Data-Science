```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(X,Y,title):
  ax = plt.gca()
  ax.axhline(y=0, color='k')
  ax.axvline(x=0, color='k')
  plt.plot(X,Y,color='green')
  plt.title(title)
  plt.ylabel('Y')
  plt.xlabel('X')
  plt.show()

def get_array():
  res = []
  start = -6
  for i in range(1,120):
    res.append(start)
    start = start + 0.1
  return res
```

**Сигмоида**

Сигмо́ида — это гладкая монотонная возрастающая нелинейная функция, имеющая форму буквы «S», которая часто применяется для «сглаживания» значений некоторой величины.
Сигмоиды применяются в нейронных сетях в качестве функций активации. Они позволяют нейронам как усиливать слабые сигналы, так и не насыщаться от сильных сигналов.
\begin{align}
σ(x)= \frac{1}{1+e^{-x}}
\end{align}

```python
def sigmoid(x):
 return 1/(1 + np.exp(-x))

X = get_array()
Y = [sigmoid(el) for el in X]
plot(X,Y,'Sigmoid activation function')
```

**ELU**

ELU — это функция, которая быстрее сводит стоимость к нулю и дает более точные результаты. В отличие от других функций активации, ELU имеет дополнительную альфа-константу, которая должна быть положительным числом.

ELU очень похож на RELU, за исключением отрицательных входных данных. Они оба находятся в форме функции тождества для неотрицательных входных данных. С другой стороны, ELU медленно сглаживается до тех пор, пока его выход не станет равным -α, тогда как RELU резко сглаживается.
\begin{align}
R(z) =
\begin{Bmatrix}
z& z > 0\\
\alpha * (e^{z}-1)& z <= 0
\end{Bmatrix}
\end{align}

```python
def elu(z,alpha):
	return z if z >= 0 else alpha*(np.exp(z) -1)

X = get_array()
Y = [elu(el,1) for el in X]
plot(X,Y,'ELU activation function')
```

RELU — это функция, обозначающая выпрямленные линейные единицы. Формула обманчиво проста: max(0,z). Несмотря на свое название и внешний вид, она не является линейной и обеспечивает те же преимущества, что и сигмоида (то есть возможность изучать нелинейные функции), но с более высокой производительностью.
\begin{align}
R(z) =
\begin{Bmatrix}
z& z > 0\\
0& z <= 0
\end{Bmatrix}
\end{align}

```python
def relu(z):
  return max(0, z)

X = get_array()
Y = [relu(el) for el in X]
plot(X,Y,'RELU activation function')
```

Масштабированные экспоненциальные линейные единицы — это функции активации, вызывающие самонормировку. Активации нейронов сети SELU автоматически сходятся к нулевому среднему и единичной дисперсии.
\begin{align}
S(z) =
\begin{Bmatrix}
\lambda*z& z > 0\\
\lambda * α(e^{x}-1)& z <= 0
\end{Bmatrix}
\end{align}

```python
def SELU(x, lambdaa = 1.0507, alpha = 1.6732):
    if x >= 0:
        return lambdaa * x
    else:
        return lambdaa * alpha * (np.exp(x) - 1)

X = get_array()
Y = [SELU(el) for el in X]
plot(X,Y,'SELU activation function')
```

SoftPlus представляет собой плавное приближение к функции ReLU и может использоваться для ограничения выходных данных машины, чтобы они всегда были положительными.

\begin{align}
S(x)= log(1+e^{x})
\end{align}

```python
def Softplus(x):
    return np.log(1 + np.exp(-np.abs(x))) + np.maximum(x,0)

X = get_array()
Y = [Softplus(el) for el in X]
plot(X,Y,'Softplus activation function')
```

В качестве альтернативы гиперболическому тангенсу softsign представляет собой функцию активации для нейронных сетей. Несмотря на то, что функции tanh и softsign тесно связаны, tanh сходится экспоненциально, тогда как softsign сходится полиномиально. Несмотря на то, что softsign появляется в литературе, на практике он не будет применяться так часто, как tanh.
\begin{align}
S(x) = \frac{x}{1+|x|}
\end{align}

```python
def soft_sign(x):
	return x/(abs(x)+1)

X = get_array()
Y = [soft_sign(el) for el in X]
plot(X,Y,'soft_sign activation function')
```

Гиперболический тангенс сжимает число с действительным знаком в диапазоне [-1, 1]. Но в отличие от Sigmoid, его выходные данные центрированы по нулю. Поэтому на практике тангенсная нелинейность всегда предпочтительнее сигмовидной нелинейности.
\begin{align}
tanh(x) = \frac{e^z-e^{-z}}{e^z+e^{-z}}
\end{align}

```python
def tanh(z):
	return (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))

X = get_array()
Y = [tanh(el) for el in X]
plot(X,Y,'tanh activation function')
```
